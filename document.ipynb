> py manage.py makemigrations

 Is a Django command used to create new migration files based on the changes you have made to your Django models. These migration files are Python files that represent the changes to your database schema.

When you run makemigrations, Django compares the current state of your models with the state of the database schema defined by previous migrations. It then creates new migration files that contain the necessary changes to bring the database schema in sync with the current state of your models.

> py manage.py migrate

This command will execute the migration files and update the database schema accordingly.

This command is used to create a superuser account within your Django project. A superuser has all permissions and can access the Django admin interface. When you run this command, Django will prompt you to enter a username, email address, and password for the superuser account. Once you provide this information, the superuser account will be created and you can use it to access the Django admin interface and perform administrative tasks.
